# Simulation of photochemical reaction kinetics under contiuous irradiation of arbitraty irradiation source


$$\frac{\mathrm d \mathbf c}{\mathrm dt}=\frac{1}{V}\int\boldsymbol\Phi(\lambda)\mathbf q(t, \lambda)\mathrm d\lambda$$



$$\frac{\mathrm d \mathbf c}{\mathrm dt}=\frac{q_{mol}^0}{V}\boldsymbol\Phi\int \frac{ \mathbf{c}\circ \boldsymbol{\varepsilon}}{\mathbf c\cdot \boldsymbol{\varepsilon}}\left(1-10^{-l\mathbf c\cdot \boldsymbol{\varepsilon}}\right)I_S(\lambda)\mathrm d\lambda$$

or in purely matrix operations

$$\frac{\mathrm d \mathbf c}{\mathrm dt}=\frac{q_{mol}^0}{V}\boldsymbol\Phi\ \mathrm{diag}\left[ \int \frac{\boldsymbol{\varepsilon}}{\mathbf c\cdot \boldsymbol{\varepsilon}}\left(1-10^{-l\mathbf c\cdot \boldsymbol{\varepsilon}}\right)I_S(\lambda)\mathrm d\lambda\right]  \mathbf{c}$$

Where $\mathbf c=
\begin{bmatrix}
     c_1(t) \\
     c_2(t) \\
     \vdots  \\
     c_n(t) 
\end{bmatrix}
$ is a concentration vector of total $n$ components and $\boldsymbol\varepsilon=
\begin{bmatrix}
     \varepsilon_1(\lambda) \\
     \varepsilon_2(\lambda) \\
     \vdots  \\
     \varepsilon_n(\lambda) 
\end{bmatrix}
$ is molar abs. coefficient vector that contains molar abs. coefficients of all $n$ components. $\mathbf K=
\begin{bmatrix}
     -\Phi_{1\rightarrow} & \Phi_{2\rightarrow1}  & \dots  & \Phi_{n\rightarrow 1} \\
     \Phi_{1\rightarrow2} & -\Phi_{2\rightarrow} & \dots  & \Phi_{n\rightarrow 2} \\
     \vdots  & \vdots & \ddots & \vdots \\
     \Phi_{1\rightarrow n} & \Phi_{2\rightarrow n}  & \dots  & -\Phi_{n\rightarrow} 
\end{bmatrix}
$ is transfer matrix. The main diagonal contains total decaying quantum yields of each compoent. Off-diagonal elements represents transition quantum yields between compartments. Finally, the $I_S(\lambda)$ is spectrum of irradiation source which is normallized $\int_{-\infty}^{+\infty}I_S(\lambda)\mathrm d\lambda=1$. In principle, Kasha rule is not followed in many photochemical reactions (eg. photoisomerizations) and therefore quantum yield depends on wavelength. In this case, the transfer matrix $\mathbf K$ is function of wavelength and should be put inside the integral. However, transfer matrix can be left outside of the integral at the cost of averaging of quantum yields for particular irradiation source. 

The simple example is cis-trans isomerization of alkene if both isomers are thermally stable. The equation for E and Z isomers would be as follow:

$$\frac{\mathrm d }{\mathrm dt}
\begin{bmatrix}
     c_{\mathrm E}(t) \\
     c_{\mathrm Z}(t) 
\end{bmatrix}
=\frac{q_{mol}^0}{V}
\begin{bmatrix}
     -\Phi_{\mathrm E\rightarrow\mathrm Z} & \Phi_{\mathrm Z\rightarrow\mathrm E} \\
     \Phi_{\mathrm E\rightarrow\mathrm Z} & -\Phi_{\mathrm Z\rightarrow\mathrm E}
\end{bmatrix}
\int \begin{bmatrix}
     c_{\mathrm E}(t)\varepsilon_{\mathrm E}(\lambda) \\
     c_{\mathrm Z}(t)\varepsilon_{\mathrm Z}(\lambda) 
\end{bmatrix}
\frac{1-10^{-l(c_{\mathrm E}(t)\varepsilon_{\mathrm E}(\lambda) + c_{\mathrm Z}(t)\varepsilon_{\mathrm Z}(\lambda))}}{c_{\mathrm E}(t)\varepsilon_{\mathrm E}(\lambda) + c_{\mathrm Z}(t)\varepsilon_{\mathrm Z}(\lambda)}I_S(\lambda)\mathrm d\lambda$$

which splits into set of two coupled ordinary differential equations with no analytical solution (if we irradiate at single wavelength $\lambda_{irr}$). The second equation is not necessary in this case, because of conservation of mass. The sum of concentrations of E and Z is constant and equal to total starting concentration. Also, this is clear from transfer matrix, because it has not full rank! In cases when $\mathbf K$ has full rank, all equations must be solved.


$$\begin{align*}
\frac{\mathrm d c_{\mathrm E}(t)}{\mathrm dt} &= \frac{q_{mol}^0}{V}\frac{1-10^{-l(c_{\mathrm E}(t)\varepsilon_{\mathrm E}(\lambda_{irr}) + c_{\mathrm Z}(t)\varepsilon_{\mathrm Z}(\lambda_{irr}))}}{c_{\mathrm E}(t)\varepsilon_{\mathrm E}(\lambda_{irr}) + c_{\mathrm Z}(t)\varepsilon_{\mathrm Z}(\lambda_{irr})}\Big( \Phi_{\mathrm Z\rightarrow\mathrm E} c_{\mathrm Z}(t)\varepsilon_{\mathrm Z}(\lambda_{irr}) - \Phi_{\mathrm E\rightarrow\mathrm Z} c_{\mathrm E}(t)\varepsilon_{\mathrm E}(\lambda_{irr})  \Big)\\
c_0 &= c_{\mathrm E}(t) + c_{\mathrm Z}(t)
\end{align*}$$



## Equilibrium of general photoreaction - reaching of the Photostationary state (PSS)

When irradiating at single wavelength, 

$$\frac{\mathrm d \mathbf c}{\mathrm dt}=\frac{1}{V}\boldsymbol\Phi\mathbf q(\mathbf c, \boldsymbol\varepsilon)$$

where $\mathbf q(\mathbf c, \boldsymbol\varepsilon)$ is the amount of light absorbed by each compratment. For the perpendicular irradiaiton setup, this can be calculated when irradiating at single wavelength $\lambda_{irr}$ or narrow band irradiation source as:

$$\mathbf q(\mathbf c, \boldsymbol\varepsilon) = \frac{\mathbf{c}\circ \boldsymbol{\varepsilon}(\lambda_{irr})}{\mathbf c\cdot \boldsymbol{\varepsilon}(\lambda_{irr})}\left(1-10^{-l\mathbf c\cdot \boldsymbol{\varepsilon}(\lambda_{irr})}\right)$$

or written with only matrix multiplications

$$\mathbf q(\mathbf c, \boldsymbol\varepsilon) = \frac{1-10^{-l\mathbf c\cdot \boldsymbol{\varepsilon}(\lambda_{irr})}}{\mathbf c\cdot \boldsymbol{\varepsilon}(\lambda_{irr})} \mathrm{diag}\left(\boldsymbol{\varepsilon}(\lambda_{irr})\right)\mathbf c$$

The term in the left is in the literature denoted as photoinetic factor $F_k(t)$ and depends only on the total absorbance of the solution at $\lambda_{irr}$

$$\mathbf q(\mathbf c, \boldsymbol\varepsilon) = F_k(t)\ \mathrm{diag}\left(\boldsymbol{\varepsilon}(\lambda_{irr})\right)\mathbf c$$

For general $n$ component reaction rewritten with vector and matrices:

$$q(\mathbf c, \boldsymbol\varepsilon) = F_k(t) \begin{bmatrix}
     \varepsilon_1(\lambda_{irr}) & 0  & \dots  & 0 \\
     0 & \varepsilon_2(\lambda_{irr}) & \dots  & 0\\
     \vdots  & \vdots & \ddots & \vdots \\
     0 & 0  & \dots  & \varepsilon_n(\lambda_{irr}) 
\end{bmatrix}
\begin{bmatrix}
     c_1(t) \\
     c_2(t) \\
     \vdots  \\
     c_n(t) 
\end{bmatrix}$$

Now, the equilibrium of a reaction is established when the concentration of all compartments does not change with time, so that $$\frac{\mathrm d \mathbf c}{\mathrm dt}=0$$

Therefore with combining above equations, this condition must hold

$$\frac{\mathrm d \mathbf c}{\mathrm dt}=\frac{F_k(t)}{V}\boldsymbol\Phi\ \mathrm{diag}\left(\boldsymbol{\varepsilon}(\lambda_{irr})\right)\mathbf c =  0$$

The relative population of all compartments in the equilibrium can be found by solving this matrix equation, which is nothing else than a **null-space** (or kernel) of a matrix.

$$\mathbf c_{PSS}  = \mathrm{Null}\left[   \boldsymbol\Phi\ \mathrm{diag}\left(\boldsymbol{\varepsilon}(\lambda_{irr})\right)\right]$$


## Analytical solution for known photokinetic factor

$$\frac{\mathrm d \mathbf c(t)}{\mathrm dt}=\frac{q_0F(t)}{V}\boldsymbol\Phi\text{diag}\left(\boldsymbol\varepsilon\right)\mathbf c(t)$$

$$\mathbf c(t) = \exp\left(\frac{q_0}{V}\boldsymbol\Phi\text{diag}\left(\boldsymbol\varepsilon\right)\int_0^tF(t^{\prime})\,\mathrm dt^{\prime}\right)\mathbf{c_0}$$

Eigenvector eigenvalue decomposition

$$\boldsymbol\Phi\text{diag}\left(\boldsymbol\varepsilon\right) = \mathbf{Q\Lambda Q^{-1}}$$

$$\mathbf c(t) = \mathbf Q\exp\left(\frac{q_0}{V}\mathbf\Lambda \int_0^tF(t^{\prime})\,\mathrm dt^{\prime}\right) \mathbf{Q^{-1}}\mathbf{c_0}$$

for wavelength dependend QY and irr with full emission source

$$\mathbf c(t) = \prod\mathbf Q\exp{\left(\frac{I_0(\lambda)}{V}\mathbf\Lambda \int_0^tF(\lambda, t^{\prime})\,\mathrm dt^{\prime}\right)} \mathbf{Q^{-1}}\ \mathbf{c_0}$$

```python
from numpy.linalg import eig, inv, svd, lstsq
from functools import partial
%matplotlib inline

@vectorize(nopython=True, fastmath=False)
def photokin_factor(A, coeff=1e-4):
    ln10 = np.log(10)
    ll2 = ln10 ** 2 / 2

    if A < coeff:
        return ln10 - A * ll2  # approximation with first two taylor series terms
    else:
        return (1 - np.exp(-A * ln10)) / A  # exact photokinetic factor

def find_nearest_idx(array, value):
    idx = np.searchsorted(array, value, side="left")
    if idx > 0 and (idx == len(array) or math.fabs(value - array[idx - 1]) < math.fabs(value - array[idx])):
        return idx - 1
    else:
        return idx
    
def simul_photokin_model(I0, c0, K=None, eps=None, times=None, V=0.003, l=1, t0=0, lambda_irr=None, R=0.036, use_backref_correction=True,
                        coeff=1e-4):
    """
    no wl dependence of K
    I0 - irradiation spectrum scalled by q0 = PDF * q0 so that integral(I0) = q0
    eps - spectra (= epsilons)
    t0 - time of start of irradiaiton
    l - length of cuvette
    V - volume in the cuvette in L
    c0 - total initial concentration vector of compounds
    """

    c0 = np.asarray(c0)

    assert c0.shape[0] == K.shape[0]
    ln10 = np.log(10)
    
    if lambda_irr is not None:
        idx = find_nearest_idx(wavelengths, lambda_irr)

    def dc_dt_spectrum(c, t):
        p_A = c[:, None] * eps * l  # hadamard product - partial absorbance
        A = p_A.sum(axis=0)  # total absorbance

        F = photokin_factor(A, coeff)  # (1-10^-A) / A
        
        Ieff = I0 * (1 - R) * (1 + R * np.exp(-A * ln10)) if use_backref_correction else I0 * (1 - R)

        product = K * (F * Ieff * eps * l).sum(axis=-1)  # K @ sum(diag(F * I0 * eps * l))

        irr_on = 1 if t >= t0 else 0

        return irr_on * product.dot(c) / V  # final dot product / V
    
    def dc_dt(c, t):
        p_A = c * eps[:, idx] * l  # hadamard product - partial absorbance
        A = p_A.sum()  # total absorbance

        F = photokin_factor(A, coeff)  # (1-10^-A) / A
        
        Ieff = I0 * (1 - R) * (1 + R * np.exp(-ln10 * A)) if use_backref_correction else I0 * (1 - R)

        product = F * Ieff * K.dot(p_A) / V  # K @ sum(diag(F * I0 * eps * l))

        irr_on = 1 if t >= t0 else 0

        return irr_on * product

    result = odeint(dc_dt_spectrum if lambda_irr is None else dc_dt, c0, times)

    return result


def simulate_C(t: np.ndarray, K: np.ndarray, j: np.ndarray) -> np.ndarray:
    Lambda, U = eig(K)  # get eigenvalues and eigenvectors of K
    U_inv = inv(U)  # calculate the inverse of U

    # calculate A_II^T matrix
    A_II_T = U * U_inv.dot(j)  # U @ diag(U^-1 @ j)
    
    # calculate basis of folded exponentials
    C = np.exp(t[:, None] * Lambda[None, :])
    
    # make linar combination of folded exponentials
    return C.dot(A_II_T.T)  # C @ A_II

def make_K(ks):
    k1, k_1, k2 = ks
    K = np.asarray([[-k1 - k2, k_1, 0], 
                    [k1, -k_1 - k2, 0],
                    [k2, k2, 0]])
    
    return K

# def make_K_example(ks):
#     k1, k2 = ks
#     K = np.asarray([[-k1, 0], 
#                     [k1, -k2]])
    
#     return K

times = np.linspace(0, 3, 50)
K = make_K([0.5, 0.33, 0.01])

# C = simulate_C(times, K, np.asarray([1, 0]))
# C = simul_photokin_model(5e-8 * LED_405, [5e-5, 0], K, ST, times)
R = 0.036
# ST_test = ST * 1e-6
q0 = 2e-7
c0 = 10 * 1e-6
# C = simul_photokin_model(q0 , [c0, 0, 0], K, ST, times, lambda_irr=415, use_backref_correction=False, R=R)
# C_scaled = simul_photokin_model(q0 * 1e6 , [c0 * 1e6, 0], K, ST * 1e-6, times, lambda_irr=415, use_backref_correction=False, R=R) * 1e-6
c0_vec = np.asarray([c0, 0, 0]) * 1e6
c0E_vec = np.asarray([0, c0, 0]) * 1e6

# C_scaled_coeff = simul_photokin_model(q0 * 1e6 , [c0 * 1e6, 0], K, ST * 1e-6, times, lambda_irr=415, use_backref_correction=False, R=R) * 1e-6

simul_abbr = partial(simul_photokin_model, q0*1e6, K=K, eps=ST * 1e-6, times=times,
                     lambda_irr=450, use_backref_correction=True, R=R)

# C = simul_photokin_model(q0 * 1e6 , c0_vec, K, ST * 1e-6, times, lambda_irr=450, use_backref_correction=True, R=R) * 1e-6
# CE = simul_photokin_model(q0 * 1e6 , c0E_vec, K, ST * 1e-6, times, lambda_irr=450, use_backref_correction=True, R=R) * 1e-6

C = simul_abbr(c0_vec) * 1e-6
CE = simul_abbr(c0E_vec) * 1e-6

plt.plot(times, C)
plt.plot(times, CE, ls='--')
# plt.plot(times, C_bc, ls='--', c='k')
plt.xlabel('Time')
plt.ylabel('Concentration')
# plt.legend(['no correction - A', 'no correction - B', 'back-reflection correction A', 'back-reflection correction B'])
plt.tight_layout()
# plt.savefig('high-conc.profiles.png', dpi=500)
plt.show()

# plt.plot(times, C_scaled - C_bc)
# plt.xlabel('Time')
# plt.ylabel('Concentration difference')

# plt.show()
```

## Model and Quantum yields


Quantum yield depends on wavelength in general and will be described as a low $n$-th order polynomial

$$\Phi_{k}(\lambda) = \phi_{k0} + \sum_{i=1}^{n}\phi_{ki}\left(\frac{\lambda_C - \lambda}{100}\right)^i$$

the parameter $\phi_{k0}$ denotes the quantum yield at wavelenght $\lambda_C$

```python
def Phi(phis, lambda_C=400, wavelengths=wavelengths):
    assert isinstance(phis, (list, np.ndarray))
    return sum(par * ((lambda_C - wavelengths) / 100) ** i for i, par in enumerate(phis))

# a = np.asarray([1, 2, 3])
# a.sum(axis=0, keepdims=True)

def _x_symlog(start, end, n, linscale=1, linthresh=1):
    """Generates n points from symlog space given start and end points
     as used in matplotlib."""
    start = np.sign(start) * linthresh * ((np.log10(np.abs(start)) - np.log10(linthresh)) / linscale + 1) if np.abs(start) >= linthresh else start
    end = np.sign(end) * linthresh * ((np.log10(np.abs(end)) - np.log10(linthresh)) / linscale + 1) if np.abs(end) >= linthresh else end
    lin = np.linspace(start, end, n)
    symlog = np.empty_like(lin)

    for i in range(lin.shape[0]):
        symlog[i] = lin[i] if np.abs(lin[i]) < linthresh else np.sign(lin[i]) * 10 ** (linscale * (np.abs(lin[i]) / linthresh - 1) + np.log10(linthresh))
    return symlog
```

```python
# batch eigenvector-eigenvalue decomposition with inverse matrix calculation
@njit(fastmath=False)
def eigen_decomp(QY_mat):
    
    n, m = QY_mat.shape[0], QY_mat.shape[1]
    
    eigs = np.zeros((n, m))
    vectors = np.zeros((n, m, m))
    vectors_inv = np.zeros((n, m, m))
    
    for i in range(QY_mat.shape[0]):
        w, wr = np.linalg.eig(QY_mat[i])
        eigs[i] = w
        vectors[i] = wr
        vectors_inv[i] = np.linalg.inv(wr)
    return eigs, vectors, vectors_inv



def simulate_an(q0, V, c0, eps, K, times, wavelengths, D, l=1, I_source=None):
    c0 = np.asarray(c0)
    
#     idx = 422-230
    
    K_aug = K * eps.T[:, None, :]  # K dot diag(eps)
    eigs, vectors, vectors_inv = eigen_decomp(K_aug)
    
    nt, nw = times.shape[0], wavelengths.shape[0]
    
    F = photokin_factor(D)
    F_integrated = np.insert(cumtrapz(F, x=times, axis=0), 0, np.zeros(nw), axis=0)
    
#     traces = np.zeros((nt, K.shape[2]))
    
#     traces = np.exp(q0 / V * F_integrated[:, idx, None] * eigs[idx])
    
#     traces = np.exp(q0 / V * F_integrated.reshape((nt, nw, 1)) * eigs[None, ...] * I_source[None, :, None])

#     for ti in prange(nt):
#         traces[ti] = vectors[idx].dot(np.diag(traces[ti])).dot(vectors_inv[idx]).dot(c0)
    
    for ti in prange(nt):
        tmp = np.exp(q0 / V * F_integrated[ti, :, None] * I_source[:, None] * eigs)
        s = np.ones_like(K[0])
        for wi in range(nw):
#             traces[ti, wi] = vectors[wi].dot(np.diag(traces[ti, wi])).dot(vectors_inv[wi]).dot(c0)
            s *= vectors[wi].dot(np.diag(tmp[wi])).dot(vectors_inv[wi])
    
        traces[ti] = s.dot(c0)

    
    return traces#.prod(axis=1).squeeze()

#     # Q @ np.diag(L) @ Q_inv

#     traces = np.exp(q0 / V * F_integrated[:, None] * L)

#     # n x n x n
#     outers = Q.T[..., None] * Q_inv[:, None, :]
#     res = traces[..., None, None] * outers[None, ...]
#     summed = res.sum(axis=1)

#     return summed.dot(c0)



traces = simulate_an(q0, 0.003, [0, c0, 0], ST, K, times, wavelengths, D, l=1, I_source=I_source)
# F = photokin_factor(D)

# traces = simulate_an(q0, 0.003, [0, c0, 0], ST, K, times, wavelengths, D, l=1, w_irr=w_irr)

plt.rcParams['figure.figsize'] = [8, 6]
# # # plt.plot(times, F)
# # # plt.plot(times, F_integrated)

for i in range(n):
    plt.plot(times, traces[:, i], label=names[i])
    plt.plot(times, C[:, i], label=names[i], color='gray', ls='--')


# plt.plot(times, F_integrated[:, 250])

# traces.shape

```
